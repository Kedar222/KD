```python
s=''
with open('/content/Test_File.txt','r') as f:
  for line in f:
    for word in line.split():
      if len(word)>len(s):
        s = word
print(s)
```

9. Write a Python program to count the frequency of words in a file.


```python
word_count={}
with open('/content/Test_File.txt','r') as f:
  for line in f:
    for word in line.split():
      if word in word_count:
        word_count[word]+=1
      else:
        word_count[word] =1

for word, count in word_count.items():
  print(f'{word} and {count}')
```
