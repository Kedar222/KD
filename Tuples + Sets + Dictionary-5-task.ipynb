###`Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]

s=[]

for i in test_list:
  s.append(i[0])

unique=set(s)

result=[]
for i in unique:
  result.append([i])
  for j in test_list:
    if j[0]==i:
      result[-1].append(j[1])
print(result)

````

###`Q2:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
a=(1, 5, 7, 8, 10)

num=[]

num.append(a[0]*a[1])

for i in range(1,len(a)-1):
  num.append(a[i]*a[i-1]+a[i]*a[i+1])

num.append(a[-1]*a[-2])

print(tuple(num))
````

###`Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
L = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

List=0
Set=0
Tuple=0
for i in L:
  if type(i)==list:
    List=List+1
  elif type(i)==set:
    Set=Set+1
  elif type(i)==tuple:
    Tuple=Tuple+1

print(f'List-{List}',f'Set-{Set}',f'Tuple-{Tuple}',sep='\n')
````

###`Q1:` Write a program to find set of common elements in three lists using sets.
```
Input : ar1 = [1, 5, 10, 20, 40, 80]
        ar2 = [6, 7, 20, 80, 100]
        ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

Output : [80, 20]
```

````python
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

ar4=[]
for i in ar1:
  for j in ar2:
    for k in ar3:
      if k==j==i:
        ar4.append(k)
print(sorted(ar4,reverse=True))
````

### `Q3`: Convert List to List of dictionaries. Given list values and keys list, convert these values to key value pairs in form of list of dictionaries.

**Example 1:**

Input:
```bash
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]
```

Output:

```bash
[{'name': 'DataScience', 'id': 3}, {'name': 'is', 'id': 8}]
```

**Example 2:**

Input:
```bash
test_list = ["CampusX", 10]
key_list = ["name", "id"]
```

Output:

```bash
[{'name': 'CampusX', 'id': 10}]
```

````python
test_list = ["CampusX", 10]
key_list = ["name", "id"]
new=[]
chunk=len(key_list)

for i in range(0,len(test_list),chunk):
  chunk=test_list[i:i+ chunk]
  new.append(dict(zip(key_list,chunk)))
print(new)
````

````python
key_list = ["name", "id"]
test_list = ["DataScience", 3, "is", 8]
result=[]
chunk_size=len(key_list)

for i in range(0, len(test_list), chunk_size):
    chunk = test_list[i : i + chunk_size]
    result.append(dict(zip(key_list, chunk)))

print(result)
````
